```python
new_file = new_file.iloc[1:,[0,1,2,3,4,9,10,11,12,13,16]]
new_file.columns = ['year','month','day','hour','minute',
                    'wave_height','wave_peak_period','wave_mean_period',
                    'wave_mean_direction',
                    'water_temperature']
new_file.index = [datetime.datetime(year,month,day,hour,minute) for year,month,day,hour,minute \
                  in new_file.iloc[:,:5].astype(int).values]
```

```python
new_file
```

```python
new_file = new_file.where((new_file!=99.0) & (new_file!=999.0) & (new_file!=9999.0))
```

```python
# new_file.index = new_file.index.round('H')
```

```python
new_file
```

```python
no_closest = ~np.isclose(new_file.wave_peak_period,new_file.wave_mean_period,atol=15)
bad_periods = new_file.wave_mean_period[no_closest].copy()
new_file.wave_mean_period.where(~no_closest,
    new_file.wave_mean_direction[no_closest].copy(),inplace=True)
new_file['wave_mean_direction_corr'] = new_file.wave_mean_direction.copy()
new_file['wave_mean_direction_corr'].where(~no_closest,bad_periods.copy(),inplace=True)
```

```python
new_file
```

```python
new_file.wave_mean_period.plot(), new_file.wave_mean_direction.plot(alpha=0.9), new_file.wave_mean_direction_corr.plot(c='red')
```

```python
new_files.append(new_file.drop(columns=['year','month','day','hour','minute']))
```
